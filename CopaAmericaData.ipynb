Event 1: Carry
- A Carry occurs when "A player controls the ball at their feet while moving or standing still"

```python
final_events_Carry.columns
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Solo para visualizacion mas linda (en canchas)
def dibujar_campo(ax=None):
    if ax is None:
        ax = plt.gca()

    # dimensiones
    ax.set_xlim(0, 120)
    ax.set_ylim(0, 80)

    # lineas
    ax.plot([0, 0], [0, 80], color="black")
    ax.plot([0, 120], [80, 80], color="black")
    ax.plot([120, 120], [80, 0], color="black")
    ax.plot([120, 0], [0, 0], color="black")
    ax.plot([60, 60], [0, 80], color="black")
    
    #  areas
    ax.plot([0, 18], [18, 18], color="black")
    ax.plot([0, 18], [62, 62], color="black")
    ax.plot([18, 18], [18, 62], color="black")
    ax.plot([120, 102], [18, 18], color="black")
    ax.plot([120, 102], [62, 62], color="black")
    ax.plot([102, 102], [18, 62], color="black")
    
    # circulo central 
    center_circle = plt.Circle((60, 40), 10, color="black", fill=False)
    ax.add_patch(center_circle)

    # punto central 
    ax.plot(60, 40, 'o', color="black")
    
    return ax


# visualizo 
def graficar_movimientos(df, equipo, periodo, ax, color):
    df_filtrado = df[(df['possession_team'] == equipo) & (df['period'] == periodo)]
    for _, row in df_filtrado.iterrows():
        start_pos = row['location']
        end_pos = row['carry_end_location']
        if start_pos and end_pos:  # No NaN data
            # Graph as an arrow 
            ax.annotate('', xy=end_pos, xytext=start_pos,
                        arrowprops=dict(facecolor=color, edgecolor=color, arrowstyle='->'))


# Analisis lo quiero dividir tanto en equipos como en tiempos (en el caso de alargue van a ser 4) 
equipos = final_events_Carry['possession_team'].unique()
periodos = final_events_Carry['period'].unique()

 
fig, axs = plt.subplots(len(equipos), len(periodos), figsize=(15, 10))

colores = {
    'Argentina': 'skyblue',  
    'Colombia': 'red'      
}

for i, equipo in enumerate(equipos):
    for j, periodo in enumerate(periodos):
        ax = axs[i, j]
        ax.set_title(f'{equipo} - Mitad {periodo} Carry')
        dibujar_campo(ax)
        color = colores.get(equipo, 'grey')  
        graficar_movimientos(final_events_Carry, equipo, periodo, ax, color)

plt.tight_layout()
plt.show()
```

- Dribble events describe when an attacker with the ball takes on another attacker.

```python
final_events_Dribble
```

```python
final_events_Dribble.columns
```

```python
#Me interesa saber cuantos dribbles exitosos tuvo cada equipo como una especie de success rate de enfrentamientos
# Suponiendo que 'final_events' es tu DataFrame que contiene los eventos de dribble

success_value = 'Complete'
fail_value = 'Incomplete'

dribble_success_count = final_events_Dribble[final_events_Dribble['dribble_outcome'] == success_value].groupby('player').size()
dribble_fail_count = final_events_Dribble[final_events_Dribble['dribble_outcome'] == fail_value].groupby('player').size()

# Creaamos un df con los resultados
dribble_stats = pd.DataFrame({
    'successful_dribbles': dribble_success_count,
    'failed_dribbles': dribble_fail_count
}).fillna(0)  

# Calculamos la tasa de éxito
dribble_stats['success_rate'] = dribble_stats['successful_dribbles'] / (dribble_stats['successful_dribbles'] + dribble_stats['failed_dribbles'])
dribble_stats['fail_rate'] = 1- dribble_stats['success_rate']

dribble_stats = dribble_stats.round({'successful_dribbles': 0, 'failed_dribbles': 0, 'success_rate': 3, 'fail_rate':3})


dribble_stats = dribble_stats.reset_index()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Define successful and unsuccessful outcomes
successful_outcomes = ['Goal']
unsuccessful_outcomes = ['Blocked', 'Off T', 'Post', 'Saved', 'Wayward', 'Saved Off T', 'Saved To Post']

# Classify each shot as successful or not successful
final_events_Shot['success'] = final_events_Shot['shot_outcome'].apply(
    lambda x: 'Successful' if x in successful_outcomes else 'Not Successful'
)

# Function to draw a football field
def draw_pitch(ax):
    # Set field dimensions
    ax.set_xlim(0, 120)
    ax.set_ylim(0, 80)

    # Draw the boundaries and center line
    ax.plot([0, 0], [0, 80], color="black")
    ax.plot([0, 120], [80, 80], color="black")
    ax.plot([120, 120], [80, 0], color="black")
    ax.plot([120, 0], [0, 0], color="black")
    ax.plot([60, 60], [0, 80], color="black")
    
    # Draw goal areas
    ax.plot([0, 18], [18, 18], color="black")
    ax.plot([0, 18], [62, 62], color="black")
    ax.plot([18, 18], [18, 62], color="black")
    ax.plot([120, 102], [18, 18], color="black")
    ax.plot([120, 102], [62, 62], color="black")
    ax.plot([102, 102], [18, 62], color="black")
    
    # Draw center circle and spot
    center_circle = plt.Circle((60, 40), 10, color="black", fill=False)
    ax.add_patch(center_circle)
    ax.plot(60, 40, 'o', color="black")

    # Draw 6-yard and 18-yard boxes
    ax.plot([0, 6], [6, 6], color="black")
    ax.plot([6, 6], [6, 14], color="black")
    ax.plot([6, 0], [14, 14], color="black")
    ax.plot([0, 6], [14, 14], color="black")
    ax.plot([120, 114], [6, 6], color="black")
    ax.plot([114, 114], [6, 14], color="black")
    ax.plot([114, 120], [14, 14], color="black")
    ax.plot([120, 114], [14, 14], color="black")

# Get unique teams and periods
teams = final_events_Shot['team'].unique()
periods = final_events_Shot['period'].unique()

# Create subplots
fig, axs = plt.subplots(len(teams), len(periods), figsize=(15, 10), sharex=True, sharey=True)

# Plot settings
colors = {'Successful': 'green', 'Not Successful': 'red'}

# Plot data
for i, team in enumerate(teams):
    for j, period in enumerate(periods):
        ax = axs[i, j]
        draw_pitch(ax)
        df_filtered = final_events_Shot[
            (final_events_Shot['team'] == team) & (final_events_Shot['period'] == period)
        ]
        
        # Plot successful shots
        df_successful = df_filtered[df_filtered['success'] == 'Successful']
        ax.scatter(df_successful['location'].apply(lambda x: x[0]), 
                   df_successful['location'].apply(lambda x: x[1]), 
                   color=colors['Successful'], label='Successful', marker='o')
        
        # Plot unsuccessful shots
        df_unsuccessful = df_filtered[df_filtered['success'] == 'Not Successful']
        ax.scatter(df_unsuccessful['location'].apply(lambda x: x[0]), 
                   df_unsuccessful['location'].apply(lambda x: x[1]), 
                   color=colors['Not Successful'], label='Not Successful', marker='x')
        
        # Set titles and labels
        ax.set_title(f'{team} - Period {period}')
        ax.legend(loc='upper right')

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Función para dibujar un campo de fútbol
def dibujar_campo(ax=None):
    if ax is None:
        ax = plt.gca()

    # Dimensiones del campo
    ax.set_xlim(0, 120)
    ax.set_ylim(0, 80)

    # Dibujar las líneas del campo
    ax.plot([0, 0], [0, 80], color="black")
    ax.plot([0, 120], [80, 80], color="black")
    ax.plot([120, 120], [80, 0], color="black")
    ax.plot([120, 0], [0, 0], color="black")
    ax.plot([60, 60], [0, 80], color="black")
    
    # Dibujar las áreas de gol
    ax.plot([0, 18], [18, 18], color="black")
    ax.plot([0, 18], [62, 62], color="black")
    ax.plot([18, 18], [18, 62], color="black")
    ax.plot([120, 102], [18, 18], color="black")
    ax.plot([120, 102], [62, 62], color="black")
    ax.plot([102, 102], [18, 62], color="black")
    
    # Dibujar el círculo central
    center_circle = plt.Circle((60, 40), 10, color="black", fill=False)
    ax.add_patch(center_circle)

    # Dibujar el punto central
    ax.plot(60, 40, 'o', color="black")
    
    return ax

# Equipos y periodos
equipos = final_events_Pressure['team'].unique()
periodos = final_events_Pressure['period'].unique()

# Crear la figura para los subplots
fig, axs = plt.subplots(len(equipos), len(periodos), figsize=(15, 10))

colores = {
    'Argentina': 'skyblue',  # Celeste
    'Colombia': 'red'        # Rojo
}

# Iterar sobre cada equipo y cada periodo
for i, equipo in enumerate(equipos):
    for j, periodo in enumerate(periodos):
        ax = axs[i, j]
        ax.set_title(f'{equipo} - Mitad {periodo} Pressure')
        dibujar_campo(ax)
        color = colores.get(equipo, 'grey')  # Color por defecto si no se encuentra en el diccionario
        
        # Filtrar los eventos de presión para el equipo y periodo actuales
        pressure_equipo = final_events_Pressure[(final_events_Pressure['team'] == equipo) & 
                                                (final_events_Pressure['period'] == periodo)]

        # Extraer las coordenadas de la presión
        x = [loc[0] for loc in pressure_equipo['location'] if loc is not None]
        y = [loc[1] for loc in pressure_equipo['location'] if loc is not None]

        # Crear el mapa de calor
        sns.kdeplot(x=x, y=y, shade=True, cmap='coolwarm', bw_adjust=.5, ax=ax)

plt.tight_layout()
plt.show()
```
