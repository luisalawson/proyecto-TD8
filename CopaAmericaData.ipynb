## Event analysis

Event 1: Carry
- A Carry occurs when "A player controls the ball at their feet while moving or standing still"

```python
final_events_Carry.columns
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Solo para visualizacion mas linda (en canchas)
def dibujar_campo(ax=None):
    if ax is None:
        ax = plt.gca()

    # dimensiones
    ax.set_xlim(0, 120)
    ax.set_ylim(0, 80)

    # lineas
    ax.plot([0, 0], [0, 80], color="black")
    ax.plot([0, 120], [80, 80], color="black")
    ax.plot([120, 120], [80, 0], color="black")
    ax.plot([120, 0], [0, 0], color="black")
    ax.plot([60, 60], [0, 80], color="black")
    
    #  areas
    ax.plot([0, 18], [18, 18], color="black")
    ax.plot([0, 18], [62, 62], color="black")
    ax.plot([18, 18], [18, 62], color="black")
    ax.plot([120, 102], [18, 18], color="black")
    ax.plot([120, 102], [62, 62], color="black")
    ax.plot([102, 102], [18, 62], color="black")
    
    # circulo central 
    center_circle = plt.Circle((60, 40), 10, color="black", fill=False)
    ax.add_patch(center_circle)

    # punto central 
    ax.plot(60, 40, 'o', color="black")
    
    return ax


# visualizo 
def graficar_movimientos(df, equipo, periodo, ax, color):
    df_filtrado = df[(df['possession_team'] == equipo) & (df['period'] == periodo)]
    for _, row in df_filtrado.iterrows():
        start_pos = row['location']
        end_pos = row['carry_end_location']
        if start_pos and end_pos:  # No NaN data
            # Graph as an arrow 
            ax.annotate('', xy=end_pos, xytext=start_pos,
                        arrowprops=dict(facecolor=color, edgecolor=color, arrowstyle='->'))


# Analisis lo quiero dividir tanto en equipos como en tiempos (en el caso de alargue van a ser 4) 
equipos = final_events_Carry['possession_team'].unique()
periodos = final_events_Carry['period'].unique()

 
fig, axs = plt.subplots(len(equipos), len(periodos), figsize=(15, 10))

colores = {
    'Argentina': 'skyblue',  
    'Colombia': 'red'      
}

for i, equipo in enumerate(equipos):
    for j, periodo in enumerate(periodos):
        ax = axs[i, j]
        ax.set_title(f'{equipo} - Mitad {periodo} Carry')
        dibujar_campo(ax)
        color = colores.get(equipo, 'grey')  
        graficar_movimientos(final_events_Carry, equipo, periodo, ax, color)

plt.tight_layout()
plt.show()
```
